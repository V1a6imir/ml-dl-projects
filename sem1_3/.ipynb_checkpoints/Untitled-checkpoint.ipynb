### 4. Модифицируйте алгоритм градиентного спука таким образом, чтобы он останавливаля, если ошибка перестает снижаться. Для этого сравнивайте, на сколько понизилась ошибка на каждом шаге и если это снижение меньше некоторой заранее заданной очень маленькой величины - выходите из цикла. Эту величину можно взять одну миллионную, например.

```python
class Model(object):
    """Модель парной линейной регрессии"""
    def __init__(self, alpha=0.001, b0=0, b1=0):
        self.b0 = 0
        self.b1 = 0
        self.alpha = alpha
        
    def predict(self, X):
        return self.b0 + self.b1 * X
    
    def check_df(self, i):
        if isinstance(i, pd.core.frame.DataFrame):
            if i.shape[1] > 1:
                print('Ошибка. DataFrame содержит больше одного столбца')
                i = None
            else:
                i = i.squeeze()
        return i        
        
    
    def error(self, X, Y):
        X = self.check_df(X)
        Y = self.check_df(Y)
        return sum(((self.predict(X) - Y)**2) / (2 * len(X)))
    
    def fit(self, X, Y, accuracy=0.001):
        X = self.check_df(X)
        Y = self.check_df(Y)
        steps, errors = [], []
        step = 0        
        while True:
            dJ0 = sum(self.predict(X) - Y) /len(X)
            dJ1 = sum((self.predict(X) - Y) * X) /len(X)
            self.b0 -= self.alpha * dJ0
            self.b1 -= self.alpha * dJ1    
            new_err = hyp.error(X, Y)
            step += 1            
            steps.append(step)
            errors.append(new_err)
            # Добалено
            try:
                if errors[-2] - errors[-1] <= accuracy:
                    break
            except IndexError:
                pass

        return steps, errors
    
```

### 5. Убедитесь, что алгоритм градиентного спуска завершается всегда - либо если ошибка перестает существенно меняться, либо при достижении максимального количества шагов.

```python
class Model(object):
    """Модель парной линейной регрессии"""
    def __init__(self, b0=0, b1=0):
        self.b0 = 0
        self.b1 = 0
        
    def predict(self, X):
        return self.b0 + self.b1 * X
    
    def check_df(self, i):
        if isinstance(i, pd.core.frame.DataFrame):
            if i.shape[1] > 1:
                print('Ошибка. DataFrame содержит больше одного столбца')
                i = None
            else:
                i = i.squeeze()
        return i        
        
    
    def error(self, X, Y):
        X = self.check_df(X)
        Y = self.check_df(Y)
        return sum(((self.predict(X) - Y)**2) / (2 * len(X)))
    
    def fit(self, X, Y, alpha=0.001, accuracy=0.001, max_steps=1000):
        X = self.check_df(X)
        Y = self.check_df(Y)
        steps, errors = [], []
        step = 0   
        
        # Добавлено
        while len(steps) <= max_steps:
            
            dJ0 = sum(self.predict(X) - Y) /len(X)
            dJ1 = sum((self.predict(X) - Y) * X) /len(X)
            self.b0 -= alpha * dJ0
            self.b1 -= alpha * dJ1    
            new_err = hyp.error(X, Y)
            step += 1            
            steps.append(step)
            errors.append(new_err)
            try:
                if errors[-2] - errors[-1] <= accuracy:
                    break
            except IndexError:
                pass
        
        return steps, errors
```

```python
hyp = Model()
steps, errors = hyp.fit(x, y, max_steps=1000)
J = hyp.error(x, y)
J
```

### 6. Реализуйте механизм адаптивной скорости обучения. Чтобы не задавать вручную параметр alpha можно отслеживать, увеличивается ли ошибка после первого шага градиентного спуска. В начале задаем скорость обучения близкой к 1. Если ошибка после одного шага увеличивается - скорость обучения надо уменьшить (например, в 2 раза). Если не увеличивается - скорость обучения можно не трогать. Еще подумайте, как после уменьшения скорости перезапускать градиентный спуск заново.

```python
class Model(object):
    """Модель парной линейной регрессии"""
    def __init__(self, b0=0, b1=0):
        self.b0 = 0
        self.b1 = 0
        
    def predict(self, X):
        return self.b0 + self.b1 * X
    
    def check_df(self, i):
        if isinstance(i, pd.core.frame.DataFrame):
            if i.shape[1] > 1:
                print('Ошибка. DataFrame содержит больше одного столбца')
                i = None
            else:
                i = i.squeeze()
        return i        
        
    
    def error(self, X, Y):
        X = self.check_df(X)
        Y = self.check_df(Y)
        return sum(((self.predict(X) - Y)**2) / (2 * len(X)))
    
    def fit(self, X, Y, alpha=1, accuracy=0.001, max_steps=1000):
        X = self.check_df(X)
        Y = self.check_df(Y)
        steps, errors = [], []
        step = 0   
        
        while len(steps) <= max_steps:    
            dJ0 = sum(self.predict(X) - Y) /len(X)
            dJ1 = sum((self.predict(X) - Y) * X) /len(X)
            self.b0 -= alpha * dJ0
            self.b1 -= alpha * dJ1    
            new_err = hyp.error(X, Y)
            step += 1            
            steps.append(step)
            errors.append(new_err)
            if step >= 2:
                if errors[-2] - errors[-1] <= accuracy:
                    break
                # Добавлено: 
                if errors[-1] > errors[-2]:
                    alpha /= 2
        return steps, errors
```

```python
# С адаптивной скоростью обучения модель обучается намного быстрее!
hyp = Model()
steps, errors = hyp.fit(x, y, max_steps=1000)
J = hyp.error(x, y)
J
```

### 7. Добавьте к классу модели метод рисования графиков. Сделайте так, чтобы он был независим от масштаба переданных данных.

```python
class Model(object):
    """Модель парной линейной регрессии"""
    def __init__(self, b0=0, b1=0):
        self.b0 = 0
        self.b1 = 0
        self.steps = []
        self.errors = []
        
    def predict(self, X):
        return self.b0 + self.b1 * X
    
    def check_df(self, i):
        if isinstance(i, pd.core.frame.DataFrame):
            if i.shape[1] > 1:
                print('Ошибка. DataFrame содержит больше одного столбца')
                i = None
            else:
                i = i.squeeze()
        return i        
        
    
    def error(self, X, Y):
        X = self.check_df(X)
        Y = self.check_df(Y)
        return sum(((self.predict(X) - Y)**2) / (2 * len(X)))
    
    def fit(self, X, Y, alpha=1, accuracy=0.001, max_steps=1000):
        X = self.check_df(X)
        Y = self.check_df(Y)
        steps, errors = [], []
        step = 0   
        
        while len(steps) <= max_steps:    
            dJ0 = sum(self.predict(X) - Y) /len(X)
            dJ1 = sum((self.predict(X) - Y) * X) /len(X)
            self.b0 -= alpha * dJ0
            self.b1 -= alpha * dJ1    
            new_err = hyp.error(X, Y)
            step += 1            
            steps.append(step)
            errors.append(new_err)
            if step >= 2:
                if errors[-2] - errors[-1] <= accuracy:
                    break
                # Добавлено: 
                if errors[-1] > errors[-2]:
                    alpha /= 2
        self.steps = steps
        self.errors = errors
        return steps, errors
    
    def plot(self):
        plt.plot(self.steps, self.errors)
    
```

```python
class Model(object):
    """Модель парной линейной регрессии"""
    def __init__(self, b0=0, b1=0):
        self.b0 = b0
        self.b1 = b1
    def predict(self, X):
        return self.b0 + self.b1 * X
    def error(self, X, Y):
        return sum(abs(self.predict(X) - Y) / (len(X)))
    def fit(self, X, Y, alpha=0.01, accuracy=0.01, max_steps=5000):
        steps, errors = [], []
        step = 0        
        for _ in range(1000):
            dJ0 = sum(self.predict(X) - Y) /len(X)
            dJ1 = sum((self.predict(X) - Y) * X) /len(X)
            self.b0 -= alpha * dJ0
            self.b1 -= alpha * dJ1    
            new_err = hyp.error(X, Y)
            step += 1            
            steps.append(step)
            errors.append(new_err)
        return steps, errors


hyp = Model()
steps, errors = hyp.fit(x, y, max_steps=1000)
J = hyp.error(x, y)

plt.figure()
plt.plot(steps, errors, 'g')
plt.show()
```

### C MAPE:

```python
class Model(object):
    """Модель парной линейной регрессии"""
    def __init__(self, b0=0, b1=0):
        self.b0 = b0
        self.b1 = b1
    def predict(self, X):
        return self.b0 + self.b1 * X
    def error(self, X, Y):
        return sum((abs(self.predict(X) - Y)/abs(Y)) / (len(X)) )
    def fit(self, X, Y, alpha=0.01, accuracy=0.001, max_steps=5000):
        steps, errors = [], []
        step = 0        
        for _ in range(1000):
            dJ0 = sum(self.predict(X) - Y) /len(X)
            dJ1 = sum((self.predict(X) - Y) * X) /len(X)
            self.b0 -= alpha * dJ0
            self.b1 -= alpha * dJ1    
            new_err = hyp.error(X, Y)
            step += 1            
            steps.append(step)
            errors.append(new_err)
        return steps, errors


hyp = Model()
steps, errors = hyp.fit(x, y, max_steps=1000)
J = hyp.error(x, y)

plt.figure()
plt.plot(steps, errors, 'g')
plt.show()
```
