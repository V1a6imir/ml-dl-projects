7. Найдите процентное соотношение учащихся, сдавших экзамен на «отлично», «хорошо», «удовлетворительно», «неудовлетворительно».

```python
grade_5 = (data[data['Балл'] >= 68].shape[0])/data.shape[0]*100
grade_4 = (data[(data['Балл'] >= 53) & (data['Балл'] < 68)].shape[0])/data.shape[0]*100 
grade_3 = (data[(data['Балл'] >= 36) & (data['Балл'] < 53)].shape[0])/data.shape[0]*100 
grade_2 = (data[(data['Балл'] >= 0) & (data['Балл'] < 36)].shape[0])/data.shape[0]*100 

print(f'Отлично {round(grade_5, 3)} %, хорошо {round(grade_4, 3)} %, удовлетворительно {round(grade_3, 3)} %, неудовлетворительно {round(grade_2, 3)} %')
```

11. Пусть задания с кратким ответом будут задания типа В. Соответственно всего по экзамену вопросов класса В: В
, ... Вк
 Посчитайте процент выполненных и невыполненных заданий по каждому вопросу класса В.

```python
B = list(data['Задания с кратким ответом'])
count = 0

for element in B[:-1]:
    for j in range(27):
        if element[j] == '-' or element[j] == '0':
            count += 1
            
summary = 0            
for element in B[:-1]:
    for j in range(27):
            summary += 1

            
print(f'Выполненые: {round((summary-count)/summary*100, 3)} %, невыполненные: {round(count/summary*100, 3)} %')
```

12. Аналогично и с типом С (ответы с развернутым ответом)

```python
C = list(data['Задания с развёрнутым ответом'])
count = 0

for element in C[:-1]:
    for j in range(20):
        if element[j] == '0':
            count += 1
            
summary = 0            
for element in C[:-1]:
    for j in range(27):
            summary += 1

            
print(f'Выполненые: {round((summary-count)/summary*100, 3)} %, невыполненные: {round(count/summary*100, 3)} %')
```

13. Сделайте анализ по двум школам:
    1. по всем выполненным заданиям типа В
    2. по заданиям типа С больше 50%
    3. по среднему баллу юношей и девушек #152 #140 

### A

```python
data_st = data.loc[(data['№ школы'] == 152) | (data['№ школы'] == 140)]
B_school = []
for el in list(data_st['Задания с кратким ответом']):
    B_school.append(round(el.count('+') / len(el) * 100, 2))
print('Процент выполнения задания В каждым учеником школ:')
print(B_school)
```

### B

```python
data_st = data.loc[(data['№ школы'] == 152) | (data['№ школы'] == 140)]
С_school = []
for el in list(data_st['Задания с развёрнутым ответом']):
    temp = 0
    for i in range(len(el)):
        if i%4 == 0: 
            if int(el[i]) >= int(int(el[i+2])/2):
                temp += 1
    С_school.append(round(temp/(len(el)/4)*100))
print('Процент выполнения заданий С больще 50% каждый учеником школ:')            
print(С_school)
```

### C

```python
data_st = data.loc[(data['№ школы'] == 152) | (data['№ школы'] == 140)]
print('Средний балл девушек школы №152 и школы №140:')
print(round(data_st[data_st['Пол']=='Ж'].mean()['Балл'], 2))
print('Средний балл юношей школы №152 и школы №140:')
print(round(data_st[data_st['Пол']=='М'].mean()['Балл'], 2))
```

14. Для каждого задания добавьте соответствующий раздел в ноутбук и текстовые пояснения.

```python
B_task = []
for el in list(data['Задания с кратким ответом']):
    temp = []
    for i in el:
        temp.append(i)
    B_task.append(temp)
    
for i in range(len(B_task)):
    B_task[i].append(list(data.index)[i])
    
B_rez = pd.DataFrame(B_task, columns = ['B1', 'B2', 'B3', 'B4', 'B5', 'B6', 'B7', 'B8', 'B9', 'B10', 'B11', 'B12', 'B13', 'B14', 'B15', 'B16', 'B17', 'B18', 'B19', 'B20', 'B20', 'B21', 'B22', 'B23', 'B24', 'B25', 'B26', 'B27'])

result = pd.concat([data, B_rez], axis=1, join='inner')
pd.set_option('display.max_columns', 200)
result.head(10)
cols = result.columns.tolist()
result = result[cols[:4] + cols[10:] + cols[6:10]]
result
```

```python
C_task = []
for el in list(data['Задания с развёрнутым ответом']):
    temp = []
    for i in range(len(el)):
        if i%4==0:
            temp.append(el[i])
            temp.append(el[i+2])
    C_task.append(temp)
    
for i in range(len(C_task)):
    C_task[i].append(list(data.index)[i])
    
C_rez = pd.DataFrame(C_task, columns = ['C1', 'C1 max', 'C2', 'C2 max', 'C3', 'C3 max', 'C4', 'C4 max', 'C5', 'C5 max', '№'])

result_end = pd.concat([result, C_rez], axis=1, join='inner')
cols = result_end.columns.tolist()
result_end = result_end[cols[:16] + cols[20:] + cols[16:20]]
result_end
```
